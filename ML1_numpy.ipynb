```python
#comparing memory occupied by both
import numpy as np
import sys
#memory occupied by list
s = range(1000)
print('memory occupied by list',sys.getsizeof(10)*len(s))
#memory occupied by numpy
d = np.arange(1000)
print('memory occupied by num array',d.size*d.itemsize)
```
