```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def slope_intercept(x1,y1,x2,y2):
  a = (y2 - y1) / (x2 - x1)
  b = y1 - a * x1
  return a,b
m,c = slope_intercept(3,5,9,10)
print(slope_intercept(3,5,9,10))
print("Equation of line : y = {0:.2f}.x + {1}".format (m,c))
```

```python
X =[171,151,124,134,156]
Y =[80,60,45,50,65]
```

```python
# calculating coefficient
mean_x = np.mean(X)
mean_y = np.mean(Y)
n = len(X)
n
```

```python
numer = 0
denom = 0
for i in range(n):
     numer += (X[i] - mean_x) * (Y[i] - mean_y)
     denom += (X[i] - mean_x) ** 2
b1 = numer / denom
b0 = mean_y - (b1 * mean_x)
print(b1,b0)
```

```python
plt.scatter(X, Y, color='red' , label='blue')
```

```python
#calculating root mean square error
rmse = 0
for i in range(n):
  y_pred = b0+b1*X[i]
  rmse +=(Y[i]-y_pred) **2
rmse = np.sqrt(rmse/n)
print("RMSE")
print(rmse)  
```

```python
from sklearn.metrics import mean_squared_error
data = pd.read_csv("/content/headbrain.csv")
data.head()

X = data['Head Size(cm^3)'].values
Y = data['Brain Weight(grams)'].values
data.iloc[0:5,0:1]
m = len(X)
X = X.reshape((m,1))

reg = LinearRegression()
reg = reg.fit(X,Y)
Y_pred = reg.predict(X)

print('Intercept: \n',reg.intercept_)
print('coefficients: \n',reg.coef_)

#model Evaluation
rmse = np.sqrt(mean_squared_error(Y, Y_pred))
r2 = reg.score(X,Y)
print("RMSE")
print(rmse)
print("R2 score")
print(r2)
```
